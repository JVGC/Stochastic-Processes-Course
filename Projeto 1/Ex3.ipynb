### Exercício 3:
    Uma moeda é lançada sucessivamente até que duas caras apareçam em sequência. Seja p a probabilidade de sair
    cara
    Simule esse problema e compare o resultado com o valor teórico. Faça um gráfico da probabilidade obtida
    em função do número de experimentos.

```python
## Variando o numero de simulações de 20 a 100 (de 20 em 20)
## Vou calcular para várias probabilidade qual o numero médio de passos dados
## E plotar uma linha para numero de simulações (n)

plt.figure(figsize=(16,8))

Nmax = 200 ## Numero máximo de simulações
vp = np.arange(0.02,1,0.02) ## probabilidade de tirar cara 

## Estilos de pontos para cada uma das simulações
aux= 0
lineStyles = ['x', '1', 'v', 's', 'o']

## Variando o Numero de Simulações
for n in np.arange(40,Nmax+1, 40):
    
    vmean = [] ## Vetor com as médias de passos
    
    ## Para cada uma das probabilidades
    for p in vp:
        vt = [] ## vetor contando o numero de passos
        for s in np.arange(0,n): ## Simulando n vezes
            t = 0
            i = 0
            ## Enquanto não conseguir 2 caras seguidas
            while (i < 2):
                ## Se tirar cara
                if(np.random.uniform() < p):
                    i = i+1
                else: ## Se tirar coroa
                    if(i>0):
                        i = i -1
                t = t +1
            vt.append(t)
        vmean.append(np.mean(vt)) ## Adiciono a média para aquela probabilidade
    ## Ploto a linha para aquele numero de simulações
    plt.plot(vp, vmean, lineStyles[aux], color='red', linewidth=2, markersize=5, markerfacecolor='red', label = str(n)+' Simulations')
    aux = aux +1

#Theoretical result
vR = []
for p in vp:
    vR.append((1+p)/(p**2)) ## Probabilidade de ruina = 1 - probabilidade de vitória
plt.plot(vp, vR, '-o', color='black', linewidth=1, markersize=5,markerfacecolor='black', label='Theoretical')


plt.xlabel("p", fontsize=20)
plt.ylabel("Numero Médio de Passos", fontsize=20)
plt.legend()
plt.grid(True)
plt.show(True) 
```
