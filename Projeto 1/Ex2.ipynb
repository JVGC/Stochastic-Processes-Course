### Exercício 2:

Para o problema da ruína do apostador, temos
$$
Ui0 = P(XT = 0|X0 = i) =
\begin{cases} 
1 - \frac{(q/p)^i - (q/p)^N}{1 - (q/p)^N} \quad p\neq q\\ 
\frac{N-i}{N} \quad p = q.
\end{cases}
$$
onde N é a fortuna total (apostador mais oponente), $p$ é a probabilidade ganho em uma jogada e $q = 1− p$.

A probabilidade do jogador A vencer em um jogo é $p = 0,49$. Ele começa o jogo com 6 reais. Seu oponente,
o jogador B, começa com 10 reais. Qual a probabilidade de A vencer? Assuma que em cada jogo é apostado 2 reais.

Simule o problema acima e compare o resultado com o resultado teórico

```python
plt.figure(figsize=(16,8))

Nmax = 1000 ## Numero máximo de Simulações
p = 0.49 ## Probabilidade de A vencer uma jogada
N = 16 ## Valor a ser obtido para ganhar o jogo
vwins = [] ## vetor para guarda a porcentagem de vitórias para cada numero de simulações

simulations = np.arange(10,Nmax+1, 5) ##  Vetor variando numero de simulações

for n in simulations:## Para cada numero de simulações
    nwins = 0 ## Calculando quantas vitórias houveram para cada numero de simulações
    for s in np.arange(0,n):
        i = 6
        while (i > 0  and i < N):
            ## Se ganhar a jogada
            if(np.random.uniform() < p):
                i = i + 2
            else: ## Se perder a jogada
                i = i - 2

        if(i == N): ## Se ele ganhou
            nwins = nwins + 1
    vwins.append(nwins/n) ## guarda a porcentagem de vitórias para aquele numero
plt.plot(simulations, vwins, color='blue', linewidth=2, label='Simulation')

## Resultado Teórico
plt.axhline(y= 0.338, color='red', linestyle='--', label='Theoretical')

plt.ylim(ymax = 2, ymin = -1)
plt.xlabel("Porcentagem de Vitórias", fontsize=20)
plt.ylabel("Numero Médio de Passos", fontsize=20)
plt.legend()
plt.show(True)
```
