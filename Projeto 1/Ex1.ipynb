```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

    função do número de experimentos.

### OBS:

    Assim:
    - Os nós variam de 0 a 4 e não de 1 a 5;
    - O nó inicial da mosca é o nó 0;
    - O nó da aranha é o nó 2, e o nó da janela é o 4

### a) Qual é a probabilidade da mosca escapar pela janela sem ser pega pela aranha? Assuma que a mosca se move pelo grafo escolhendo uma aresta com a mesma probabilidade.

```python
plt.figure(figsize=(16,8))

vp = np.array([(1,3), (0,2), (2, ), (0,2,4), (4, )]) ## Vetor de Probabilidades
Nmax = 1000 # Numero máximo de Simulações
simulations = np.arange(10,Nmax+1, 5) ##  Vetor variando numero de simulações
vwin = [] ## vetor para guarda a porcentagem de absorção para cada numero de simulações
for n in simulations: ## Para cada numero de simulações
    nwins = 0 ## Calculando quantas absorçoes houveram para cada numero de simulações
    
    for s in np.arange(0,n): ## Simulando n vezes
        t = 0
        i = 0
        ## Enquanto nao for absorvido
        while((i != 4) and (i != 2)):
            ## Escolhe um nó entre os possíveis
            i = random.choice(vp[i])
        if(i == 4): ## Se for o no 4, soma uma absorção
            nwins = nwins + 1
    vwin.append(nwins/n) ## guarda a porcentagem de absorções para aquele numero
    

plt.plot(simulations, vwin, color='blue', linewidth=2, label='Simulation')

## Resultado Teórico
plt.axhline(y=2/7, color='red', linestyle='--', label='Theoretical')

plt.ylim(ymax = 1, ymin = -0)
plt.xlabel("Numero de Simulações", fontsize=20)
plt.ylabel("Numero Médio de Passos", fontsize=20)
plt.legend()
plt.show(True)
```

### b) Calcule o número médio de visitas às posições 2 e 4 antes da mosca ser pega pela aranha ou sair pela janela. Assuma que a mosca iniciou na posição 1.

```python
plt.figure(figsize=(16,8))

vp = np.array([(1,3), (0,2), (2, ), (0,2,4), (4, )]) ## Vetor de Probabilidades
Nmax = 1000 # Numero máximo de Simulações
simulations = np.arange(10,Nmax+1, 5) ##  Vetor variando numero de simulações
vmean = [] ## vetor para guarda a média de passo para cada numero de simulações

for n in simulations: ## Para cada numero de simulações
    vt = [] ## vetor para guardar o numero de passos dados nos nós 1 e 3 naquela tentativa
    
    for s in np.arange(0,n):## Simulando n vezes
        t = 0 ## numero de passos
        i = 0 ## no
        ## Enquanto nao for absorvido
        while((i != 4) and (i != 2)):
            i = random.choice(vp[i])
            ## Se o no for o 1 ou o 3, somo um passo dado
            if((i == 1) or (i == 3)):
                t = t +1
        vt.append(t)
    vmean.append(np.mean(vt)) ## guarda a média de passos para aquele numero
plt.plot(simulations, vmean, color='blue', linewidth=2, label='Simulation')

## Resultado Teórico
plt.axhline(y=12/7, color='red', linestyle='--', label='Theoretical')

plt.ylim(ymax = 3, ymin = 0)
plt.xlabel("Numero de Simulações", fontsize=20)
plt.ylabel("Numero Médio de Passos", fontsize=20)
plt.legend()
plt.show(True)
```

### c) Assuma agora que a aranha não está na posição 3. Qual é o tempo médio que a mosca leva para sair pela janela?

```python
plt.figure(figsize=(16,8))

vp = np.array([(1,3), (0,2), (1,3), (0,2,4), (4, )]) ## Vetor de Probabilidades
Nmax = 1000 # Numero máximo de Simulações
simulations = np.arange(10,Nmax+1, 5) ##  Vetor variando numero de simulações
vmean = [] ## vetor para guarda a média de passo para cada numero de simulações

for n in simulations: ## Para cada numero de simulações
    vt = [] ## vetor para guardar o numero de passos dados naquela tentativa
    
    for s in np.arange(0,n): ## Simulando n vezes
        t = 0 ## numero de passos
        i = 0 ## no
        
        ## Enquanto não for absorvido
        while(i != 4):
            ## Escolhe um nó entre os possíveis
            i = random.choice(vp[i])
            t = t +1
        vt.append(t)
    vmean.append(np.mean(vt))  ## guarda a média de passos para aquele numero
plt.plot(simulations, vmean, color='blue', linewidth=2, label='Simulation')

## Resultado Teórico
plt.axhline(y=12, color='red', linestyle='--', label='Theoretical')

plt.ylim(ymax = 20, ymin = 4)
plt.xlabel("Numero de Simulações", fontsize=20)
plt.ylabel("Numero Médio de Passos", fontsize=20)
plt.legend()
plt.show(True) 
```
